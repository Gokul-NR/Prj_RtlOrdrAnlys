```python
bad_strings = ['NA', 'N/A', 'na', 'nan', 'unknown', 'Unknown',
               'NOT AVAILABLE', 'Not Available', 'missing', 'Missing', '']

# Count per column
for col in df.columns:
    count = df[col].isin(bad_strings).sum()
    if count > 0:
        print(f"{col}: {count}")
```

```python
mask = df.isin(bad_strings)
rows_with_bad_values = df[mask.any(axis=1)]
print(rows_with_bad_values)
```

```python
bad_strings = ['na', 'n/a', 'nan', 'unknown', 'not available', 'missing', '']

bad_strings_lower = [x.lower() for x in bad_strings]

for col in df.columns:
    # Convert to string, strip spaces, lowercase
    clean_series = df[col].astype(str).str.strip().str.lower()
    count = clean_series.isin(bad_strings_lower).sum()
    if count > 0:
        print(f"{col}: {count}")
```

```python
keywords = ['na', 'n/a', 'nan', 'unknown', 'not available', 'missing']

for col in df.columns:
    clean_series = df[col].astype(str).str.strip().str.lower()

    # Check partial matches
    mask_partial = clean_series.apply(
        lambda x: any(keyword in x for keyword in keywords)
    )

    # Check empty strings
    mask_empty = clean_series == ''

    total_count = mask_partial.sum() + mask_empty.sum()

    if total_count > 0:
        print(f"{col}: {total_count} problematic values")
```

```python
print(df['State Name'].unique())
print(df['Dist Name'].unique())
```

```python
bad_strings = [
    'na', 'n/a', 'nan', 'unknown', 'not available', 'missing', ''
]
import numpy as np

bad_set = set(bad_strings)

for col in ['State Name', 'Dist Name']:
    df[col] = df[col].astype(str).str.strip()
    df[col] = df[col].apply(
        lambda x: np.nan if x.strip().lower() in bad_set else x
    )

```

```python
for col in df.columns:
    df[col] = df[col].astype(str).str.strip()
    df[col] = df[col].apply(
        lambda x: np.nan if x.strip().lower() in bad_set else x
    )
```

```python
print(df['State Name'].isnull().sum())
print(df['Dist Name'].isnull().sum())
```

```python
for col in df.columns:
    count = (df[col].astype(str).str.strip() == '').sum()
    if count > 0:
        print(f"{col}: {count} empty strings")
```

```python
for col in df.columns:
    uniques = df[col].unique()
    if len(uniques) < 10:
        print(f"{col}: {uniques}")
```

```python
tolerance = 1e-8

# Create a boolean DataFrame
zero_mask = np.isclose(df, 0.0, atol=tolerance)

# Count zeros in each column
print(zero_mask.sum())
```

```python
# Try to convert every column to numeric, keep non-numeric as-is
df_numeric = df.apply(pd.to_numeric, errors='coerce')

# Now safely check for near-zeros
zero_counts = np.isclose(df_numeric, 0.0, atol=1e-8).sum()
print(zero_counts)
```

```python
zero_counts_per_column = np.isclose(df_numeric, 0.0, atol=1e-8).sum(axis=0)
print(zero_counts_per_column)
```

```python
# Get series instead of array
zero_counts_per_column = pd.Series(
    np.isclose(df_numeric, 0.0, atol=1e-8).sum(axis=0),
    index=df.columns
)

# Print nicely
print(zero_counts_per_column)
```
